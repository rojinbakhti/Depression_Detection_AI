## This block reads individual egemaps files and converts them to csv files and concats them into a dataframe and put them in df.


```python
# ids_to_read = all_ids[0:1]
# name = "vgg16"
def read_and_save_features(name,ids_to_read,combined_name):
  df=pd.DataFrame()
  count = 0
  for i in ids_to_read:
    count += 1
    with tarfile.open(f"AVEC_Challenge_Packed/data/{i}_P.tar") as tar:
      subdir_and_files = [
      tarinfo for tarinfo in tar.getmembers()
      if tarinfo.name.startswith(f"{i}_P/features/{i}_{name}.csv")
      ]

      member = subdir_and_files[0]
      #csv_file = io.StringIO(tar.extractfile(member).read().decode('ascii'))
      csv_file = io.TextIOWrapper(tar.extractfile(member), encoding="utf-8")
      curr_df= csv.reader(csv_file)
      #curr_df = pd.read_csv(csv_file, delimiter=';')  #this line is used for egemaps, mfcc,
      curr_df = pd.read_csv(csv_file) #this is used for vgg16
      curr_df['patient_id'] = i 
      curr_df.to_csv(f"extracted_features/{i}_{name}.csv")
      df = pd.concat([df, curr_df], ignore_index=False, axis=0)
      print(f"{count}/{len(ids_to_read)}: Saving {i}_{name}.csv")
  df.to_csv(f'{name}_{combined_name}.csv')
  print(f"End: Saving {name}_{combined_name}.csv")
  #curr_df.head()
  del df
```
